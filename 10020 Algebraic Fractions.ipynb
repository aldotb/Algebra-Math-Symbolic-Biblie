```python
from Libaldomath  import *
```

```python
P=MyEq((x*x-10*x+13)/((x-1)*(x-2)*(x-3)),'P')
```

```python
P2=MyEq(m/(x-1)+n/(x-2)+p/(x-3),'P2')
```

```python
Q=MQ(P,P2)
```

```python
Q.factor('R')
```

```python
Q.setL(numer(Q.L))
```

```python
Q.setR(numer(Q.R))
```

```python
p1,p2=Q.coef_list()
```

```python
Q1=MQ(p1[0],p2[0])
Q2=MQ(p1[1],p2[1])
Q3=MQ(p1[2],p2[2])
```

```python
m=Q1.solve('m')
```

```python
Q2.set(m)
```

```python
Q3.set(m)
```

```python
n=Q2.solve('n')
```

```python
Q3.set(n)
```

```python
p=Q3.solve('p')
```

```python
n.set(p)
```

```python
m.set(p,n)
```
