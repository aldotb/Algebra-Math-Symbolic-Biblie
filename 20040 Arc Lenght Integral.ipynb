```python
from  libaldomath  import *
from lib_MyIntegral import *
```

```python
Q=MQ((y+1)**2,4*x**3)
```

```python
y=Q.solve(y)
```

```python
def Larc(obj):
    obj=obj2expr(obj)
    obj=diff(obj,x)
    expr=sqrt(1+obj*obj)
    return expr
```

```python
Lx=Larc(y)
Lx
```

```python
L=MyIntg(Lx,'L',var=x,x1=0,x2=3)
```

```python
L.doit()
```
